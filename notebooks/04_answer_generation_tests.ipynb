# 04 — Answer Generation Tests

Run the full QA pipeline: question → retrieval → LLM → answer + sources. Uses project config for index and eval paths.

**Prerequisites:** Indexing pipeline run; `OPENAI_API_KEY` set in `.env`. If the key is missing or retrieval score is below `RETRIEVAL_MIN_SCORE`, the pipeline returns a fallback message (no LLM call).

**Contents:** Full QA with two questions, then retrieval-only mode (no LLM) to inspect what chunks would be sent to the model.

```python
import os
from src.pipeline.qa_pipeline import run_qa

if not os.getenv("OPENAI_API_KEY"):
    print("Warning: OPENAI_API_KEY not set. Answers may be fallback-only if retrieval score is low.")

out = run_qa("What is the cancellation policy?", top_k=3)
print("Answer:", out.get("answer", ""))
print("Sources:", out.get("sources", []))
```

```python
# Second question + formatted sources
out2 = run_qa("What are the onboarding steps?", top_k=3)
print("Question: What are the onboarding steps?")
print("Answer:", out2.get("answer", ""))
print("Sources:")
for s in out2.get("sources", []):
    print(f"  - {s.get('document_name', '')} (page {s.get('page_number', '')})")
```

```python
# Retrieval-only mode (no LLM): see which chunks would be used for an answer
# Useful when OPENAI_API_KEY is not set or to debug retrieval
out_retrieval_only = run_qa("What is P1 incident response time?", top_k=3, use_llm=False)
print("Answer (fallback when use_llm=False):", out_retrieval_only.get("answer", ""))
print("Sources (chunks that would be sent to LLM):")
for s in out_retrieval_only.get("sources", []):
    print(f"  - {s.get('document_name')} (page {s.get('page_number')})")
```
