# 02 — Chunking Experiments

Inspect chunking output. Columns: `chunk_id`, `document_name`, `page_number`, `text`, `char_count`. Paths use project config (CWD-independent).

**Prerequisites:** Run extraction then chunking: `python -m src.ingestion.run_extraction` then `python -m src.ingestion.run_chunking`

**What you get:** Chunk table, chunk size distribution, chunks per document, and a sample chunk.

```python
import pandas as pd

from src.config import CHUNK_OUTPUT

if not CHUNK_OUTPUT.exists():
    raise FileNotFoundError(
        "Chunks not found. Run: python -m src.ingestion.run_extraction then python -m src.ingestion.run_chunking"
    )
df = pd.read_csv(CHUNK_OUTPUT)
print("Shape:", df.shape)
print("Columns:", list(df.columns))
df.head()
```

```python
# Chunk size distribution (char_count) and chunks per document
print("Chunk size (char_count) — min/mean/max:")
print(df["char_count"].describe()[["min", "mean", "max"]].round(1))
print("\nChunks per document:")
print(df.groupby("document_name").size().sort_values(ascending=False))
```

```python
# Sample chunk: first chunk of first document
row = df.iloc[0]
print(f"chunk_id: {row['chunk_id']}  document: {row['document_name']}  page: {row['page_number']}  char_count: {row['char_count']}")
print("-" * 50)
print((row["text"] or "")[:500])
print("...")
```
