# 03 — Retrieval Tests

Test semantic retrieval (FAISS) with sample questions. Uses project config for index paths.

**Prerequisites:** Run in order: `run_extraction` → `run_chunking` → `run_indexing_pipeline` (so the FAISS index exists).

**Score:** Higher = more semantically similar to the question (normalized similarity, typically 0–1). Scores below `RETRIEVAL_MIN_SCORE` (if set) lead to fallback answers in the QA pipeline.

```python
from src.pipeline.retrieval_pipeline import run_retrieval

# Single query
question = "What is the cancellation policy?"
results = run_retrieval(question, top_k=3)
print(f"Question: {question}\nTop-{len(results)} chunks:")
for r in results:
    doc, score, text = r.get("document_name"), r.get("score"), (r.get("text") or "")[:100]
    print(f"  {doc} (score={score:.4f}) — {text}...")
    print("  ---")
```

```python
# Multiple example queries
for q in [
    "What is the cancellation policy?",
    "What are the onboarding steps?",
    "What is P1 incident response time?",
]:
    results = run_retrieval(q, top_k=1)
    top = results[0] if results else {}
    print(f"Q: {q}")
    print(f"  Top doc: {top.get('document_name', '—')} (score={top.get('score', 0):.4f})")
    print()
```

```python
# Compare top-1 vs top-5 for one question (see how rank and score change)
q = "What is the cancellation policy?"
for k in [1, 3, 5]:
    results = run_retrieval(q, top_k=k)
    print(f"Top-{k}: {[(r.get('document_name'), round(r.get('score', 0), 4)) for r in results]}")
```
